# Importing Libraries

```python
import pandas as pd
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

### Setting Configurations

```python
pd.set_option('display.max_columns', None)
```

## Loading the dataset
- Using dataset on CO2 from ***OCO-2 MIP Top-down COâ‚‚ Budgets***

```python
data = pd.read_csv('/content/climate_nasa.csv')
```

```python
data
```

### Basic Data Cleaning

```python
data.info()
```

- Removing the mean value rows from years as it will cause issue while converting the column into numerice dtype

```python
df = data[data['Year']!= 'mean']
df.shape
```

- Dropping null values as they will not help us in visulaizations

```python
df = df.dropna()
df.shape
```

- Checking and converting the dtypes of the columns

```python
df.dtypes
```

```python
df['Year'] = pd.to_numeric(df['Year'])
df.info()
```

### Saving the dataset

```python
df.to_csv('cleaned_data.csv')
```

### The dataset is clear now. So, we can use it for visualization

- Performing Groupby on countries column to get the mean values

```python
df_mean = df.groupby('Alpha 3 Code').mean().reset_index()
```

### Creating visualizations to understand the data better

## Visualization by Country code:
- Tells us about average fossil fuel Emission by each country
- Dark part shows that emission is greater from that country

```python
# Create a choropleth map for Fossil Fuel Emissions
fig = px.choropleth(
    df_mean,
    locations='Alpha 3 Code',
    color='FF (TgCO2)',
    hover_name='Alpha 3 Code',
    color_continuous_scale='Reds',
    title='Average Fossil Fuel and Cement Emissions by Country (2015-2020)',
    labels={'FF (TgCO2)': 'Fossil Fuel Emissions (TgCO2)'}
)

# Update layout for better appearance
fig.update_layout(
    geo=dict(showframe=False, showcoastlines=True, projection_type='equirectangular'),
    coloraxis_colorbar=dict(title="Fossil Fuel Emissions<br>(TgCOâ‚‚)")
)

fig.show()
```

### Top 10 Countries with most CO2 emitters

```python
# Identify the top 10 emitters
top_emitters = df_mean.sort_values(by='FF (TgCO2)', ascending=False).head(10)

# Create a bar chart
fig = px.bar(
    top_emitters,
    x='Alpha 3 Code',
    y='FF (TgCO2)',
    title='Top 10 Countries by Average Fossil Fuel and Cement Emissions (2015-2020)',
    labels={'FF (TgCO2)': 'Fossil Fuel Emissions (TgCOâ‚‚)', 'Alpha 3 Code': 'Country'},
    text='FF (TgCO2)'
)

# Enhance the bar chart
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')

fig.show()
```

## Breakdown of COâ‚‚ Emissions by Source for Countries

# Four main Emissions breakdown


1.   Fossil Fuel Emissions: Total COâ‚‚ emissions from burning fossil fuels and cement production.
2.   Net Biosphere Exchange: Net carbon exchange between the atmosphere and biosphere, indicating absorption or release of COâ‚‚.
3.   Net Carbon Exchange: Overall balance of carbon entering and leaving a region through natural and human-induced processes.
4.   Land Carbon Stock Loss: Loss of carbon from terrestrial ecosystems due to deforestation or land-use changes.



```python
import plotly.graph_objects as go

# Define emission types and their corresponding columns
emission_types = {
    'Fossil Fuel Emissions': 'FF (TgCO2)',
    'Net Biosphere Exchange': 'LNLG NBE (TgCO2)',
    'Net Carbon Exchange': 'LNLG NCE (TgCO2)',
    'Land Carbon Stock Loss': 'IS dC_loss (TgCO2)'
}

# Create figure
fig = go.Figure()

for name, column in emission_types.items():
    fig.add_trace(go.Choropleth(
        locations=df_mean['Alpha 3 Code'],
        z=df_mean[column],
        hovertemplate='<b>%{location}</b><br>' +
                      f'{name}: %{{z}}<extra></extra>',
        colorscale='Reds',
        colorbar_title=name,
        visible=(name == 'Fossil Fuel Emissions'),
        marker_line_color='black',
        marker_line_width=0.5,
        name=name
    ))

# Create buttons for each emission type
buttons = []
for i, name in enumerate(emission_types.keys()):
    visibility = [False] * len(emission_types)
    visibility[i] = True
    buttons.append(dict(label=name,
                        method='update',
                        args=[{'visible': visibility},
                              {'title': f'{name} by Country (2015-2020)'}]))

# Update layout with buttons placed above the graph
fig.update_layout(
    updatemenus=[dict(
        type='buttons',
        direction='left',
        buttons=buttons,
        pad={'r': 10, 't': 10},
        showactive=True,
        x=0.0,          # Align buttons to the left
        xanchor='left',
        y=1.15,         # Position buttons above the plot area
        yanchor='top'
    )],
    title='Fossil Fuel Emissions by Country (2015-2020)',
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type='equirectangular'
    ),
    margin=dict(t=150)  # Increase top margin to accommodate buttons
)

fig.show()
```

### Selecting the columns to perform year wise trend on emissions and biosphere exchange

```python
# Select relevant columns
selected_columns = [
    'Year',
    'FF (TgCO2)', 'FF unc (TgCO2)',
    'LNLG NBE (TgCO2)', 'LNLG NBE unc (TgCO2)',
    'IS dC_loss (TgCO2)', 'IS dC_loss unc (TgCO2)',
    'Wood+Crop (TgCO2)', 'Wood+Crop unc (TgCO2)'
]

df_selected = df[selected_columns].copy()

# Rename columns for easier access
df_selected.rename(columns={
    'FF (TgCO2)': 'Fossil_Fuel_Emissions',
    'FF unc (TgCO2)': 'Fossil_Fuel_Uncertainty',
    'LNLG NBE (TgCO2)': 'Net_Biosphere_Exchange',
    'LNLG NBE unc (TgCO2)': 'Net_Biosphere_Uncertainty',
    'IS dC_loss (TgCO2)': 'Land_Carbon_Loss',
    'IS dC_loss unc (TgCO2)': 'Land_Carbon_Uncertainty',
    'Wood+Crop (TgCO2)': 'Wood_Crop_Emissions',
    'Wood+Crop unc (TgCO2)': 'Wood_Crop_Uncertainty'
}, inplace=True)

# Display the first few rows
df_selected.head()
```

```python
df_yearly = df_selected.groupby('Year').mean().reset_index()
# Initialize the figure
fig = go.Figure()

# Add traces for each metric
fig.add_trace(go.Scatter(
    x=df_yearly['Year'],
    y=df_yearly['Fossil_Fuel_Emissions'],
    mode='lines+markers',
    name='Fossil Fuel Emissions',
    line=dict(color='firebrick', width=2)
))

fig.add_trace(go.Scatter(
    x=df_yearly['Year'],
    y=df_yearly['Net_Biosphere_Exchange'],
    mode='lines+markers',
    name='Net Biosphere Exchange',
    line=dict(color='green', width=2)
))

fig.add_trace(go.Scatter(
    x=df_yearly['Year'],
    y=df_yearly['Wood_Crop_Emissions'],
    mode='lines+markers',
    name='Wood & Crop Emissions',
    line=dict(color='blue', width=2)
))

# Update layout
fig.update_layout(
    title='Annual Trends in COâ‚‚ Emissions and Biosphere Exchange (2015-2020)',
    xaxis_title='Year',
    yaxis_title='COâ‚‚ Emissions (TgCOâ‚‚)',
    legend_title='Metrics',
    hovermode='x unified'
)

fig.show()
```

```python
from plotly.subplots import make_subplots

# Create a subplot with 2 rows
fig = make_subplots(
    rows=2, cols=1,
    shared_xaxes=True,
    vertical_spacing=0.15,
    subplot_titles=("Fossil Fuel Emissions with Uncertainty", "Net Biosphere Exchange and Wood & Crop Emissions")
)

# Add bar chart for Fossil Fuel Emissions with error bars
fig.add_trace(
    go.Bar(
        x=df_yearly['Year'],
        y=df_yearly['Fossil_Fuel_Emissions'],
        name='Fossil Fuel Emissions',
        marker_color='firebrick',
        error_y=dict(type='data', array=df_yearly['Fossil_Fuel_Uncertainty'], visible=True)
    ),
    row=1, col=1
)

# Add line charts for NBE and Wood & Crop Emissions
fig.add_trace(
    go.Scatter(
        x=df_yearly['Year'],
        y=df_yearly['Net_Biosphere_Exchange'],
        mode='lines+markers',
        name='Net Biosphere Exchange',
        line=dict(color='green', width=2)
    ),
    row=2, col=1
)

fig.add_trace(
    go.Scatter(
        x=df_yearly['Year'],
        y=df_yearly['Wood_Crop_Emissions'],
        mode='lines+markers',
        name='Wood & Crop Emissions',
        line=dict(color='blue', width=2)
    ),
    row=2, col=1
)

# Update layout
fig.update_layout(
    title_text='Annual COâ‚‚ Metrics Overview (2015-2020)',
    showlegend=True,
    hovermode='x unified'
)

# Update y-axes titles
fig.update_yaxes(title_text="Fossil Fuel Emissions (TgCOâ‚‚)", row=1, col=1)
fig.update_yaxes(title_text="COâ‚‚ Metrics (TgCOâ‚‚)", row=2, col=1)

fig.show()
```
